```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")

credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies.shape
```

```python
credits.shape
```

```python
movies = movies.merge(
    credits,
    on='title'
)
```

```python
movies.shape
```

```python
movies = movies[
[
    'movie_id',
    'title',
    'overview',
    'genres',
    'keywords',
    'cast',
    'crew'
]
]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.shape
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['genres'].head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['keywords'].head()
```

```python
def convert3(text):
    L = []
    counter = 0

    for i in ast.literal_eval(text):

        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break

    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies['cast'].head()
```

```python
def fetch_director(text):
    
    L = []
    
    for i in ast.literal_eval(text):
        
        if i['job'] == 'Director':
            
            L.append(i['name'])
            
            break
            
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['crew'].head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview'].head()
```

```python
movies['genres'] = movies['genres'].apply(
    lambda x:[i.replace(" ","") for i in x]
)

movies['keywords'] = movies['keywords'].apply(
    lambda x:[i.replace(" ","") for i in x]
)

movies['cast'] = movies['cast'].apply(
    lambda x:[i.replace(" ","") for i in x]
)

movies['crew'] = movies['crew'].apply(
    lambda x:[i.replace(" ","") for i in x]
)
```

```python
movies['tags'] = (
    movies['overview']
    + movies['genres']
    + movies['keywords']
    + movies['cast']
    + movies['crew']
)
```

```python
movies['tags'].head()
```

```python
new_df = movies[
    ['movie_id','title','tags']
].copy()
```

```python
new_df.head()
```

```python
new_df['tags'].head()
```

```python
new_df['tags'] = new_df['tags'].apply(
    lambda x:" ".join(x)
)
```

```python
new_df['tags'] = new_df['tags'].apply(
    lambda x:x.lower()
)
```

```python
new_df['tags'].head()
```

```python
type(new_df['tags'].iloc[0])
```

```python
new_df['tags'].iloc[0]
```

```python
new_df['tags'] = new_df['tags'].apply(
    lambda x: x.lower()
)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(
    max_features=5000,
    stop_words='english'
)

vectors = cv.fit_transform(
    new_df['tags']
).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
def recommend(movie):

    movie = movie.lower()

    matching_movies = new_df[
        new_df['title'].str.lower() == movie
    ]

    if matching_movies.empty:
        print("Movie not found")
        return

    movie_index = matching_movies.index[0]

    distances = similarity[movie_index]

    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    print("\nRecommended Movies:\n")

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Avatar")
```

```python
recommend("Batman Begins")
```

```python
movie_name = input(
    "Enter Movie Name: "
)

recommend(movie_name)
```
